```python
import pandas as pd
import numpy as np

data = {
    "StudentID": ["S001","S002","S003","S004","S005","S006","S007","S008","S009","S010"],
    "Name": ["Arjun Mehta","Divya Sharma","Karan Patel","Meera Nair","Sahil Gupta",
             "Anjali Singh","Rohan Verma","Pooja Yadav","Vivek Kumar","Nisha Agarwal"],
    "Age": [21,22,20,23,21,22,20,23,21,22],
    "City": ["Delhi","Mumbai","Pune","Chennai","Delhi","Jaipur","Mumbai","Chennai","Pune","Delhi"],
    "Department": ["CSE","ECE","CSE","MECH","ECE","CSE","MECH","CSE","ECE","MECH"],
    "Score": [78,91,55,42,88,35,67,92,71,58],
    "Grade": ["B","A","C","D","A","F","B","A","B","C"],
    "Passed": [True,True,True,True,True,False,True,True,True,True]
}

df = pd.DataFrame(data)
```
