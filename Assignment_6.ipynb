```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
months = ['Jan','Feb','Mar','Apr','May','Jun',
          'Jul','Aug','Sep','Oct','Nov','Dec']

profit = [13000,17000,18000,21000,20000,25000,
          27000,25000,29000,31000,34000,37000]

expenses = [32000,35000,30000,40000,38000,42000,
            45000,44000,46000,50000,54000,58000]

revenue = [45000,52000,48000,61000,58000,67000,
           72000,69000,75000,81000,88000,95000]

df = pd.DataFrame({
    'Month': months,
    'Revenue': revenue,
    'Expenses': expenses,
    'Profit': profit
})

df
```

# Q 1.

```python
plt.figure(figsize=(10,5))

plt.plot(df['Month'],
         df['Revenue'],
         linewidth=2,
         marker='o',
         color='steelblue')

plt.title("Monthly Revenue")

plt.xlabel("Month")

plt.ylabel("Revenue")

plt.grid(alpha=0.3)

plt.show()
```

# Q 2.

```python
plt.figure(figsize=(12,6))

plt.plot(df['Month'],
         df['Revenue'],
         color='blue',
         marker='o',
         label='Revenue')

plt.plot(df['Month'],
         df['Expenses'],
         color='red',
         marker='s',
         label='Expenses')

plt.plot(df['Month'],
         df['Profit'],
         color='green',
         marker='^',
         label='Profit')

plt.title("Revenue vs Expenses vs Profit")

plt.xlabel("Month")

plt.ylabel("Amount")

plt.legend()

plt.grid(alpha=0.3)

plt.show()
```

# Q 3.

```python
plt.figure(figsize=(10,5))

plt.fill_between(df['Month'],
                 df['Revenue'],
                 alpha=0.25,
                 color='skyblue')

plt.plot(df['Month'],
         df['Revenue'],
         color='blue',
         linewidth=2)

plt.title("Revenue Area Chart")

plt.xlabel("Month")

plt.ylabel("Revenue")

plt.grid(alpha=0.3)

plt.show()
```

# Q 4.

```python
plt.figure(figsize=(10,5))

plt.fill_between(df['Month'],
                 0,
                 df['Revenue'],
                 color='skyblue',
                 alpha=0.5,
                 label='Revenue')

plt.fill_between(df['Month'],
                 0,
                 df['Expenses'],
                 color='orange',
                 alpha=0.5,
                 label='Expenses')

plt.title("Revenue and Expenses Area")

plt.xlabel("Month")

plt.ylabel("Amount")

plt.legend()

plt.show()
```

# Q 5.

```python
np.random.seed(10)

ClassA = np.random.normal(70,10,50)

ClassB = np.random.normal(80,8,50)

ClassC = np.random.normal(65,12,50)

plt.figure(figsize=(8,5))

plt.boxplot([ClassA,ClassB,ClassC],
            labels=['Class A','Class B','Class C'],
            patch_artist=True)

plt.title("Test Scores")

plt.ylabel("Marks")

plt.show()
```

# Q 6.

## Observations

- Class B has the highest median score because its median line appears higher than the other classes.
- Class C shows greater variability because its box and whiskers are wider.
- Any points outside the whiskers are considered outliers.
- The class with the highest number of visible outlier points has the most outliers.

# Q 7.

```python
plt.figure(figsize=(12,6))

plt.plot(df['Month'],
         df['Revenue'],
         color='blue',
         marker='o',
         label='Revenue')

plt.plot(df['Month'],
         df['Expenses'],
         color='red',
         marker='o',
         label='Expenses')

plt.plot(df['Month'],
         df['Profit'],
         color='green',
         marker='o',
         label='Profit')

highest = max(revenue)

index = revenue.index(highest)

plt.annotate(
    'Highest Revenue',
    xy=(months[index], highest),
    xytext=(months[index], highest+8000),
    arrowprops=dict(facecolor='black')
)

plt.axvspan(6,8,
            color='yellow',
            alpha=0.3)

plt.legend()

plt.title("Customized Revenue Chart")

plt.grid(alpha=0.3)

plt.show()
```
