# **Ribbon Plot**

```python
import plotly.graph_objects as go
import pandas as pd

nodes = [
    "Lawyers & Lobbyists", "Other", "Defense", "Communic/Electronics",
    "Transportation", "Construction", "Finance/Insur/RealEst", "Health",
    "Ideology/Single-Issue", "Agribusiness", "Misc Business", "Energy/Nat Resources",
    "Democrats", "Republicans", "Outside Spending Grps"
]

data = [
    [0, 12, 50], [0, 13, 20], [0, 14, 5],
    [1, 12, 45], [1, 13, 25], [1, 14, 10],
    [2, 12, 15], [2, 13, 35], [2, 14, 8],
    [3, 12, 30], [3, 13, 15], [3, 14, 12],
    [4, 12, 12], [4, 13, 22], [4, 14, 4],
    [5, 12, 10], [5, 13, 28], [5, 14, 6],
    [6, 12, 60], [6, 13, 75], [6, 14, 30],
    [7, 12, 40], [7, 13, 30], [7, 14, 5],
    [8, 12, 55], [8, 13, 40], [8, 14, 80],
    [9, 12, 8],  [9, 13, 24], [9, 14, 3],
    [10, 12, 20], [10, 13, 30], [10, 14, 10],
    [11, 12, 12], [11, 13, 48], [11, 14, 15]
]

df = pd.DataFrame(data, columns=['source', 'target', 'value'])

node_colors = ["#7f7f7f"] * 12 + ["#3366cc", "#ff6666", "#ffcc00"]

def get_link_color(target_index):
    if target_index == 12:   return "rgba(51, 102, 204, 0.4)"
    elif target_index == 13: return "rgba(255, 102, 102, 0.4)"
    else:                    return "rgba(255, 204, 0, 0.4)"

df['link_color'] = df['target'].apply(get_link_color)

fig = go.Figure(data=[go.Sankey(
    node = dict(
      pad = 15,
      thickness = 20,
      line = dict(color = "black", width = 0.5),
      label = nodes,
      color = node_colors
    ),
    link = dict(
      source = df['source'].tolist(),
      target = df['target'].tolist(),
      value = df['value'].tolist(),
      color = df['link_color'].tolist()
  ))])
fig.update_layout(
    title_text="Contributions During 2018 Election Cycle<br><sub>(by Sector)</sub>",
    title_x=0.5,
    font_size=12,
    width=700,
    height=800,
    plot_bgcolor='white',
    paper_bgcolor='white'
)
fig.show()
```

```python
import plotly.express as px
import pandas as pd

data = {
    "Total": ["2018 Election Cycle"] * 36,
    "Recipient": [
        "Democrats", "Democrats", "Democrats", "Democrats", "Democrats", "Democrats",
        "Democrats", "Democrats", "Democrats", "Democrats", "Democrats", "Democrats",
        "Republicans", "Republicans", "Republicans", "Republicans", "Republicans", "Republicans",
        "Republicans", "Republicans", "Republicans", "Republicans", "Republicans", "Republicans",
        "Outside Spending Grps", "Outside Spending Grps", "Outside Spending Grps", "Outside Spending Grps",
        "Outside Spending Grps", "Outside Spending Grps", "Outside Spending Grps", "Outside Spending Grps",
        "Outside Spending Grps", "Outside Spending Grps", "Outside Spending Grps", "Outside Spending Grps"
    ],
    "Sector": [
        "Lawyers & Lobbyists", "Other", "Defense", "Communic/Electronics", "Transportation", "Construction", "Finance/Insur/RealEst", "Health", "Ideology/Single-Issue", "Agribusiness", "Misc Business", "Energy/Nat Resources",
        "Lawyers & Lobbyists", "Other", "Defense", "Communic/Electronics", "Transportation", "Construction", "Finance/Insur/RealEst", "Health", "Ideology/Single-Issue", "Agribusiness", "Misc Business", "Energy/Nat Resources",
        "Lawyers & Lobbyists", "Other", "Defense", "Communic/Electronics", "Transportation", "Construction", "Finance/Insur/RealEst", "Health", "Ideology/Single-Issue", "Agribusiness", "Misc Business", "Energy/Nat Resources"
    ],
    "Amount (Millions)": [
        50, 45, 15, 30, 12, 10, 60, 40, 55, 8, 20, 12,
        20, 25, 35, 15, 22, 28, 75, 30, 40, 24, 30, 48,
        5, 10, 8, 12, 4, 6, 30, 5, 80, 3, 10, 15
    ]
}

df = pd.DataFrame(data)

color_map = {
    "2018 Election Cycle": "#d3d3d3",
    "Democrats": "#3366cc",
    "Republicans": "#ff6666",
    "Outside Spending Grps": "#ffcc00"
}

fig = px.sunburst(
    df,
    path=["Total", "Recipient", "Sector"],
    values="Amount (Millions)",
    color="Recipient",
    color_discrete_map=color_map,
    title="Contributions During 2018 Election Cycle (Hierarchical Structure)"
)

fig.update_layout(
    title_x=0.5,
    width=800,
    height=800,
    font_family="Arial"
)

fig.update_traces(
    textinfo="label+value"
)

fig.show()
```

```python
import plotly.express as px
import pandas as pd

# Using the 'df' DataFrame from the previous cell, which contains the data in a suitable format
# df was created in cell 13mIBaa-mIh0

fig = px.treemap(
    df,
    path=["Recipient", "Sector"],
    values="Amount (Millions)",
    color="Recipient",
    color_discrete_map=color_map, # Reusing the color_map from the sunburst chart
    title="Contributions During 2018 Election Cycle (Treemap by Recipient and Sector)"
)

fig.update_layout(
    title_x=0.5,
    width=800,
    height=800,
    font_family="Arial"
)

fig.update_traces(
    textinfo="label+value"
)

fig.show()
```

# **Grouped** **Bar** **Chart**



```python
fig = px.bar(
    df,
    x='Sector',
    y='Amount (Millions)',
    color='Recipient',
    barmode='group', # This creates grouped bars
    color_discrete_map=color_map, # Reusing the color_map
    title='Contributions by Sector and Recipient (Grouped Bar Chart)'
)

fig.update_layout(
    title_x=0.5,
    width=900,
    height=600,
    font_family="Arial",
    xaxis_title="Sector",
    yaxis_title="Amount (Millions)"
)

fig.show()
```

# **Stacked Bar Chart**

```python
fig = px.bar(
    df,
    x='Sector',
    y='Amount (Millions)',
    color='Recipient',
    barmode='stack', # This creates stacked bars
    color_discrete_map=color_map, # Reusing the color_map
    title='Total Sector Contributions (Stacked Bar Chart by Recipient)'
)

fig.update_layout(
    title_x=0.5,
    width=900,
    height=600,
    font_family="Arial",
    xaxis_title="Sector",
    yaxis_title="Amount (Millions)"
)

fig.show()
```

# **Pie** **Chart**

```python
df_recipient_totals = df.groupby('Recipient')['Amount (Millions)'].sum().reset_index()

fig = px.pie(
    df_recipient_totals,
    names='Recipient',
    values='Amount (Millions)',
    color='Recipient',
    color_discrete_map=color_map, # Reusing the color_map
    title='Overall Contributions by Recipient (Pie Chart)'
)

fig.update_layout(
    title_x=0.5,
    width=800,
    height=600,
    font_family="Arial"
)

fig.update_traces(
    textinfo="percent+label",
    pull=[0.05 if r == 'Outside Spending Grps' else 0 for r in df_recipient_totals['Recipient']] # Slightly pull out 'Outside Spending Grps' for emphasis
)

fig.show()
```

# **Pie Chart**

```python
df_sector_totals = df.groupby('Sector')['Amount (Millions)'].sum().reset_index()

fig = px.pie(
    df_sector_totals,
    names='Sector',
    values='Amount (Millions)',
    title='Overall Contributions by Sector (Pie Chart)'
)

fig.update_layout(
    title_x=0.5,
    width=800,
    height=600,
    font_family="Arial"
)

fig.update_traces(
    textinfo="percent+label"
)

fig.show()
```

# **Horizontal Bar Chart**

```python
fig = px.bar(
    df_recipient_totals.sort_values(by='Amount (Millions)', ascending=True),
    x='Amount (Millions)',
    y='Recipient',
    color='Recipient',
    orientation='h',
    color_discrete_map=color_map,
    title='Total Contributions per Recipient (Horizontal Bar Chart)'
)

fig.update_layout(
    title_x=0.5,
    width=800,
    height=500,
    font_family="Arial",
    xaxis_title="Amount (Millions)",
    yaxis_title="Recipient"
)

fig.show()
```

# **Horizontal Bar Chart**

```python
fig = px.bar(
    df_sector_totals.sort_values(by='Amount (Millions)', ascending=True),
    x='Amount (Millions)',
    y='Sector',
    color='Sector',
    orientation='h',
    title='Total Contributions per Sector (Horizontal Bar Chart)'
)

fig.update_layout(
    title_x=0.5,
    width=800,
    height=700,
    font_family="Arial",
    xaxis_title="Amount (Millions)",
    yaxis_title="Sector"
)

fig.show()
```
